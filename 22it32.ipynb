Python code to create a simple python calculator

```python
# Simple Python Calculator

# Function definitions
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "Error! Division by zero."
    else:
        return x / y

# Main program
print("Select operation:")
print("1. Add")
print("2. Subtract")
print("3. Multiply")
print("4. Divide")

# Take input from the user
choice = input("Enter choice (1/2/3/4): ")

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

if choice == '1':
    print(f"{num1} + {num2} = {add(num1, num2)}")
elif choice == '2':
    print(f"{num1} - {num2} = {subtract(num1, num2)}")
elif choice == '3':
    print(f"{num1} * {num2} = {multiply(num1, num2)}")
elif choice == '4':
    print(f"{num1} / {num2} = {divide(num1, num2)}")
else:
    print("Invalid input")

```

python code to check if string is palindrome

```python
def is_palindrome(s):
    s = s.lower().replace(" ", "")  # Optional: make case-insensitive and ignore spaces
    return s == s[::-1]

# Example usage
word = "madam"
if is_palindrome(word):
    print(f"'{word}' is a palindrome.")
else:
    print(f"'{word}' is not a palindrome.")
```

Python code to open display, resize, rotate, perform age detection on an image

```python
# Download prototxt
!wget -O age_deploy.prototxt https://github.com/ChandrikaDeb/Face-Detection-Age-Gender/blob/master/model/age_deploy.prototxt?raw=true

# Download caffemodel
!wget -O age_net.caffemodel https://github.com/ChandrikaDeb/Face-Detection-Age-Gender/blob/master/model/age_net.caffemodel?raw=true

import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import files

# Upload an image (face should be visible)
uploaded = files.upload()
image_path = list(uploaded.keys())[0]

# Read and display original image
img = cv2.imread(image_path)
cv2_imshow(img)

# Resize
resized_img = cv2.resize(img, (300, 300))
cv2_imshow(resized_img)

# Rotate
rotated_img = cv2.rotate(resized_img, cv2.ROTATE_90_CLOCKWISE)
cv2_imshow(rotated_img)

# Load the model
age_net = cv2.dnn.readNetFromCaffe('age_deploy.prototxt', 'age_net.caffemodel')

# Age ranges
AGE_BUCKETS = ['(0-2)', '(4-6)', '(8-12)', '(15-20)',
               '(25-32)', '(38-43)', '(48-53)', '(60-100)']

# Detect face
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(gray, 1.1, 5)

for (x, y, w, h) in faces:
    face = img[y:y+h, x:x+w].copy()

    blob = cv2.dnn.blobFromImage(face, 1.0, (227, 227),
                                 (78.4263377603, 87.7689143744, 114.895847746), swapRB=False)
    age_net.setInput(blob)
    preds = age_net.forward()
    age = AGE_BUCKETS[preds[0].argmax()]

    # Draw results
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
    cv2.putText(img, f'Age: {age}', (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX,
                0.8, (255, 0, 0), 2, cv2.LINE_AA)

cv2_imshow(img)
```

Python code to create an image slideshow

```python
from google.colab import files
uploaded = files.upload()

import time
from IPython.display import display, clear_output
from PIL import Image

# Get uploaded filenames
image_files = list(uploaded.keys())

# Loop to show each image one by one
for filename in image_files:
    img = Image.open(filename)
    clear_output(wait=True)     # Clear previous image
    display(img)                # Show current image
    time.sleep(2)               # Wait 2 seconds before next

```

Python code to extract frames from a video

```python
!pip install opencv-python-headless

from google.colab import files
uploaded = files.upload()

import cv2
import os

# Get uploaded video filename
video_path = list(uploaded.keys())[0]

# Create a folder to save frames
os.makedirs("frames", exist_ok=True)

# Read the video
cap = cv2.VideoCapture(video_path)
frame_num = 0

while cap.isOpened():
    success, frame = cap.read()
    if not success:
        break

    # Save each frame as an image
    frame_filename = f"frames/frame_{frame_num:04d}.jpg"
    cv2.imwrite(frame_filename, frame)
    frame_num += 1

cap.release()
print(f" {frame_num} frames extracted to the 'frames/' folder.")

from IPython.display import Image, display

# Display first 5 frames
for i in range(5):
    display(Image(filename=f"frames/frame_{i:04d}.jpg"))

```

Python code to AI application using python
Image Classification with AI (using a pre-trained model)

```python
!pip install tensorflow

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.applications.mobilenet import MobileNet, preprocess_input, decode_predictions
from tensorflow.keras.preprocessing import image
from google.colab import files

uploaded = files.upload()
img_path = list(uploaded.keys())[0]

img = image.load_img(img_path, target_size=(224, 224))
img_array = image.img_to_array(img)
img_batch = np.expand_dims(img_array, axis=0)
img_preprocessed = preprocess_input(img_batch)

plt.imshow(img)
plt.axis('off')
plt.title("Uploaded Image")
plt.show()

model = MobileNet(weights='imagenet')  # Load pre-trained MobileNet model

preds = model.predict(img_preprocessed)
results = decode_predictions(preds, top=3)[0]

print(" Top Predictions:")
for i, (imagenetID, label, prob) in enumerate(results):
    print(f"{i+1}. {label}: {prob*100:.2f}%")

```
